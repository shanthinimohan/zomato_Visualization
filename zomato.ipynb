```python
data = pd.merge(df, df_code, on='Country Code', how='inner')
```

```python
data.info()
```

```python
exchange_rates = {
    'Botswana Pula(P)': 6.18,
    'Brazilian Real(R$)': 17.08,
    'Dollar($)': 83.12,
    'Emirati Diram(AED)': 22.53,
    'Indian Rupees(Rs.)': 1.00,  
    'Indonesian Rupiah(IDR)': 0.0053,
    'NewZealand($)': 50.10,
    'Pounds(專)': 103.83,
    'Qatari Rial(QR)': 22.83,
    'Rand(R)': 4.55,
    'Sri Lankan Rupee(LKR)': 0.25,
    'Turkish Lira(TL)': 2.89,
}
                        
# Function to convert cost to INR
def convert_to_inr(row):
    currency = row['Currency']
    exchange_rate = exchange_rates.get(currency, 1.0)  # Default to 1 if currency not found
    return row['Average Cost for two'] * exchange_rate

# Create a new column 'Cost in INR'
data['Cost in INR'] = data.apply(convert_to_inr, axis=1)

# Display the updated DataFrame
data.columns
```

```python
data
```

```python
import plotly.express as px

# Create a scatter plot
fig = px.scatter(data, x='Currency', y=['Average Cost for two', 'Cost in INR'],
                 labels={'value': 'Cost'},
                 title='Comparison of Average Cost for two and Cost in INR',
                 color_discrete_map={'Average Cost for two': 'blue', 'Cost in INR': 'orange'})

# Increase the size of the plot and set y-axis range
fig.update_layout(height=600, width=1000, yaxis=dict(range=[0, 100000]))
# Show the plot
fig.show()
```
